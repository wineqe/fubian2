```python
import numpy as np
import matplotlib.pyplot as plt


# 
rho = 1.225  # 
U = 10  # 
a = 1  # 
theta = np.linspace(0, 2 * np.pi, 100)  # 

# 
Gamma_list = [-2 * np.pi * a * U, 0, 2 * np.pi * a * U]


def calculate_pressure(Gamma):
    # 
    # 
    V = U * (1 - a ** 2 / (a * np.exp(1j * theta)) ** 2) + 1j * Gamma / (2 * np.pi * a * np.exp(1j * theta))
    V_magnitude = np.abs(V)
    # 
    p = 0.5 * rho * U ** 2 - 0.5 * rho * V_magnitude ** 2
    # 
    Cp = (p - 0.5 * rho * U ** 2) / (0.5 * rho * U ** 2)
    return p, Cp


for Gamma in Gamma_list:
    p, Cp = calculate_pressure(Gamma)
    plt.plot(theta, Cp, label=f'Gamma={Gamma}')

plt.xlabel('theta')
plt.ylabel('Cp')
plt.title('Pressure Coefficient Distribution for Different Gamma')
plt.legend()
plt.show()
```
